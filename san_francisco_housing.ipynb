# Housing Rental Analysis for San Francisco

```python
# Import the required libraries and dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
import geoviews as gv
```

## Import the data 

```python
# Importing the sfo_neighborhoods_census_data.csv
sfo_data_df = pd.read_csv(
    Path('Resources/sfo_neighborhoods_census_data.csv')   
).drop_duplicates()

# Review the first and last five rows of the DataFrame
display(sfo_data_df.head())
display(sfo_data_df.tail())
```

---

## Calculate and Plot the Housing Units per Year

```python
# Create a numerical aggregation that groups the data by the year and then averages the results.
housing_units_by_year = (
    sfo_data_df
    .groupby('year')
    .mean()
)

display(housing_units_by_year)
```

```python
# Create a visual aggregation explore the housing units by year
housing_units_by_year.hvplot.bar(
    x='year',
    y='housing_units',
    ylim=(365000, 385000),
    title='Housing Units in San Francisco 2010 - 2016',
    ).opts(
    yformatter='%.0f')
```

The overall trend in housing_units over the period being analyzed is growth. Every year from 2010 to 2016 there are more housing units in San Francisco. There is not a year when there was a loss of units or even a maintenence of units, there is consistent growth year after year.

---

## Calculate and Plot the Average Sale Prices per Square Foot

```python
# Create a numerical aggregation by grouping the data by year and averaging the results
prices_square_foot_by_year = (
    sfo_data_df
    .groupby('year')
    .mean()
)

display(prices_square_foot_by_year)
```

The lowest gross rent reported for the years included in the DataFrame is $1239 in  2010

```python
# Filter out the housing_units column, creating a new DataFrame 
# Keep only sale_price_sqr_foot and gross_rent averages per year
prices_square_foot_by_year = (
    sfo_data_df.drop(columns=['housing_units'])
    .groupby('year')
    .mean()
)

display(prices_square_foot_by_year)
```

```python
# Plot prices_square_foot_by_year. 
prices_square_foot_by_year.hvplot(
    title = 'Sale Price per Sq Ft and Avg Gross Rent - San Francisco 2010 - 2016',
    xlabel = 'Year',
    ylabel = 'Gross Rent/Sale Price per Sq Foot'
)
```

The average sale price per square foot in 2011 was 27.440924 LESS than the average sale price per square foot in 2010.

Despite the drop in sale price per square foot in 2011, the gross rent in 2011 actually increased over the 2010 average gross rent.

---

## Compare the Average Sale Prices by Neighborhood

```python
# Group by year and neighborhood and then create a new dataframe of the mean values
prices_by_year_by_neighborhood = (
    sfo_data_df
    .groupby(['year','neighborhood'])
    .mean()
)

display(prices_by_year_by_neighborhood)
```

```python
# Filter out the housing_units
prices_by_year_by_neighborhood = prices_by_year_by_neighborhood.drop(columns=['housing_units'])

display(prices_by_year_by_neighborhood.head())
display(prices_by_year_by_neighborhood.tail())
```

```python
# Create an interactive line plot of the average price per square foot

prices_by_year_by_neighborhood.hvplot(
    title = 'Sale Price per Sq Ft & Avg Gross Rent - SFO 2010 - 2016 - by Neighborhood',
    x = 'year', 
    xlabel = 'Year',
    ylabel = 'Gross Rent/Sale Price per Sq Foot',
    groupby = 'neighborhood'
)
```

For the Anza Vista neighborhood, the average sale price per square foot for 2016 was 88.402, 256.089 less than the price for 2012 which was 344.491

## Build an Interactive Neighborhood Map

```python
# Load neighborhoods coordinates data
neighborhood_locations_df = pd.read_csv(
    Path('Resources/neighborhoods_coordinates.csv'),
    index_col = 'Neighborhood'
)
# Review the DataFrame
display(neighborhood_locations_df)
```

```python
# Calculate the mean values for each neighborhood
all_neighborhood_info_df = (
    sfo_data_df
    .groupby('neighborhood')
    .mean()
)
# Review the resulting DataFrame
all_neighborhood_info_df
```

```python
# Join the neighborhood_locations_df and the all_neighborhood_info_df DataFrame
# The axis of the concatenation is "columns".

all_neighborhoods_df = pd.concat(
    [neighborhood_locations_df, all_neighborhood_info_df], 
    axis="columns",
    sort=False
)

# Review the resulting DataFrame
display(all_neighborhoods_df.head())
display(all_neighborhoods_df.tail())
```

```python
# Call the dropna function to remove any neighborhoods that do not have data
all_neighborhoods_df = all_neighborhoods_df.reset_index().dropna()

# Rename the "index" column as "Neighborhood" for use in the Visualization
all_neighborhoods_df = all_neighborhoods_df.rename(columns={"index": "Neighborhood"})

# Review the resulting DataFrame
display(all_neighborhoods_df.head())
display(all_neighborhoods_df.tail())
```

### Using hvPlot with GeoViews enabled, create a `points` plot for the `all_neighborhoods_df` DataFrame.

```python
# Create a plot to analyze neighborhood info
all_neighborhoods_df.hvplot.points(
    'Lon',
    'Lat',
    geo = True,
    size = 'sale_price_sqr_foot',
    color = 'gross_rent',
    frame_width = 700,
    frame_height = 500,
    title = 'San Francisco Neighborhoods',
    tiles = True,
    hover_cols = 'Neighborhood'
)
```

The neighborhood has the highest gross rent appears to be Westwood Park. The neighborhood with the highest sale price per square foot appears to be Merced Heights.
